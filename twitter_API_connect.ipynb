```python
import os
import tweepy

def twitter_auth():
    consumer_key = os.environ.get('twitter_consumer_key')
    consumer_secret = os.environ.get('twitter_secret_key')
    access_token = os.environ.get('twitter_access_token')
    access_token_secret = os.environ.get('twitter_access_token_secret')
    
    auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_token, access_token_secret)
    api = tweepy.API(auth)
    
    return api
    
api = twitter_auth()
```

```python
import pandas as pd

elon_tweets = api.user_timeline('elonmusk')
elon_tweets_list = [[tweet.id for tweet in elon_tweets],
                    [tweet.text for tweet in elon_tweets],
                    [tweet.created_at for tweet in elon_tweets]]

# elon_tweets_list
elon_tweets_df = pd.DataFrame(elon_tweets_list).transpose()
elon_tweets_df.columns = ['tweet_id','tweet_text', 'tweet_datetime']
elon_tweets_df                        
```
